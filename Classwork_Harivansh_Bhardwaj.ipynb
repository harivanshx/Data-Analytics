```python
# find pokemon with maximum base_tatal and if tie exist check if legendry is true
maxbasetotal = df[df['base_total'] == df['base_total'].max()]

# if tie exist then check if legendry is true
maxbasetotal = maxbasetotal[maxbasetotal['is_legendary'] == True]
#still tie exit then check generation = 3
maxbasetotal = maxbasetotal[maxbasetotal['generation'] == 3]



maxbasetotal.head()
```

```python
max_base = df['base_total'].max()

# Find Pokémon with maximum base_total, and if a tie exists:
#   → Prefer legendary Pokémon
#   → If tie still exists, prefer generation 3
maxbasetotal = df[
    (df['base_total'] == max_base) &
    ((df['is_legendary'] == True) | (df['is_legendary'] == df['is_legendary'])) &
    ((df['generation'] == 3) | (df['generation'] == df['generation']))
]
print(maxbasetotal.head())
```
